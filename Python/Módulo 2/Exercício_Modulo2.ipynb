## 1\. Listas

Criei uma lista chamada `filmes` com o nome dos 10 primeiros filmes mais bem avaliados no site no [IMDB](https://www.imdb.com/chart/top/). Imprima o resultado.

```python
filmes = ['Um Sonho de Liberdade (1994)', 
          'O Poderoso Chefão (1972)', 
          'Batman: O Cavaleiro das Trevas (2008)', 
          'O Poderoso Chefão II (1974)', 
          '12 Homens e uma Sentença', 
          'A Lista de Schindler',
          'O Senhor dos Anéis: O Retorno do Rei (2003)',
          'Pulp Fiction: Tempo de Violência (1994)',
          'O Senhor dos Anéis: A Sociedade do Anel (2001)',
          'Três Homens em Conflito (1966)']
print(filmes)
print(type(filmes))
print(len(filmes))
```

Simule a movimentação do *ranking*. Utilize os métodos `insert` e `pop` para trocar a posição do primeiro e do segundo filme da lista. Imprima o resultado.

```python
print("--- Primeira Lista ---")
print(filmes)

primeira_posicao = filmes.pop(0)
filmes.insert(1, primeira_posicao)

print("\n--- Segunda Lista ---")
print(filmes)
print(len(filmes))
```

## 2\. Conjuntos

Aconteceu um erro no seu *ranking*. Simule a duplicação dos três últimos filmes da lista. Imprima o resultado.

```python
print(filmes)
print(type(filmes))
print(len(filmes))


print('---Filmes Duplicado no final---')
filmes.append('O Senhor dos Anéis: A Sociedade do Anel (2001)')
filmes.append('Três Homens em Conflito (1966)')
print(filmes)
print(len(filmes))





```

Utiliza a conversão `set` e `list` para remover os valores duplicados. Imprima o resultado. 

```python
print(filmes)
print(type(filmes))
print(len(filmes))

print('---Remover o elemento pelo valor com metodo remove---')
filmes.remove('O Senhor dos Anéis: A Sociedade do Anel (2001)')
filmes.remove('Três Homens em Conflito (1966)')
print(filmes)
print(len(filmes))
print(type(filmes))

print('---Removendo pela conversão set---')
converter_lista = set(filmes)
print(converter_lista)
print(len(converter_lista))
```
