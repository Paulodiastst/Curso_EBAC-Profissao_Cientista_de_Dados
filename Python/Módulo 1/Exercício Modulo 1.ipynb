## 2\. Números

Preencha as células de código para preencher os valores de (A), (B) e (C) na tabela de ticket médio abaixo:

<br>

| Dia   | Valor Total Vendas | Qtd Total Vendas | Ticket Medio |
|-------|--------------------|------------------|--------------|
| 19/01 | (A)                | 3                | 320.52       |
| 20/01 | 834.47             | (B)              | 119.21       |
| 23/01 | 15378.12           | 5                | (C)          |

```python
vtv_a = 320.52 * 3
tkm_a = vtv_a / 3

print(f'Valor Total de Vendas A: R$ {vtv_a:.2f}')
print(f'Ticket Médio A: R$ {tkm_a}')

 
```

```python
vtv_b = 834.47
qtv_b = vtv_b / 119.21

tkm_b = vtv_b / qtv_b


print(f'Valor Total de Vendas B: R$ {vtv_b:.2f}')
print(f'Ticket Médio B: R$ {tkm_b}')
```

```python
vtv_c = 15378.12 / 5
tkm_c = vtv_c / 5

print(f'Valor Total de Vendas C: R$ {vtv_c: .2f}')
print(f'Ticket Médio C: R$ {tkm_c: .2f}')
```

## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

cancao_posicao = cancao.find('roda pião')
cancao_subst = cancao.replace('roda gigante', 'roda pequena')

print(cancao.upper())
print(cancao_posicao)
print(cancao_subst)
```
