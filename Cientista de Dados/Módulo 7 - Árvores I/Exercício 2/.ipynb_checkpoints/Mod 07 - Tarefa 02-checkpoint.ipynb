- Separe 70% da base para treinamento e 30% para validação. Cada uma dessas partes terá dois objetos, um armazenando a variável resposta ```mau```e outro armazenando as variáveis explicativas (lembrando: sem variáveis string, já com as dummies).

````python
# Ler o arquivo CSV
df = pd.read_csv('https://raw.githubusercontent.com/Paulodiastst/Curso_EBAC-Profissao_Cientista_de_Dados/main/Cientista%20de%20Dados/M%C3%B3dulo%207%20-%20%C3%81rvores%20I/Exerc%C3%ADcio%201/demo01_dummies.csv')
print(df.info())
df
````

````python
# Separe 70% da base para treinamento e 30% para validação. 
# Cada uma dessas partes terá dois objetos, 
# um armazenando a variável resposta mau e outro armazenando as variáveis explicativas 
# (lembrando: sem variáveis string, já com as dummies).

# X = Variáveis explicativas:
X = df.drop(columns='mau')
print('Quantidade de linhas e colunas das variáveis explicativas (X):', 
      X.shape, '\n')

# y = Variável resposta:
y = df.mau
print('Quantidade de linhas da variável resposta (y):', 
      len(y), '\n\n')

# Separando a base em 70% para treinamento e 30% para validação:
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.3, 
                                                    random_state=100)

print('Quantidade de linhas e colunas de X_train (70%):', 
      X_train.shape)
print('Quantidade de linhas e colunas de X_test (30%):', 
      X_test.shape, '\n')

print('Quantidade de linhas de y_train (70%):', 
      len(y_train))
print('Quantidade de linhas de y_test (30%):', 
      len(y_test))
````

````python
def visualizar_arvore(decision_tree, X_columns, class_names=['bons', 'maus'], figsize=(12, 12)):
    plt.figure(figsize=figsize)
    plot_tree(decision_tree=decision_tree,
              feature_names=X_columns.tolist(),  # Convertendo o Index para lista
              class_names=class_names,
              filled=True, 
              rounded=True)
    plt.title("Árvore de Decisão")
    plt.xlabel("Atributos")
    plt.ylabel("Classes")
    plt.show()

# Suponha que clf seja seu classificador de árvore de decisão
# Suponha que X_train seja seu conjunto de dados de treinamento
# Suponha que X_train.columns seja os nomes das colunas de X_train
# Suponha que 'bons' e 'maus' sejam as classes que você quer nomear na visualização
visualizar_arvore(clf, X_train.columns)

````

````python
def visualizar_matriz_confusao(clf, X, y_true, display_labels=['aprovados', 'reprovados']):
    # Faça as previsões
    y_pred = clf.predict(X)
    
    # Calcule a matriz de confusão
    matriz_confusao = confusion_matrix(y_true, y_pred)
    
    # Plote a matriz de confusão
    plt.figure(figsize=(8, 6))
    sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=display_labels, yticklabels=display_labels)
    plt.title("Matriz de Confusão")
    plt.xlabel("Valores Preditos")
    plt.ylabel("Valores Reais")
    plt.show()

# Suponha que clf seja seu classificador de árvore de decisão
# Suponha que X_train e y_train sejam seus conjuntos de dados de treinamento
visualizar_matriz_confusao(clf, X_train, y_train)
````

````python
def visualizar_matriz_confusao(clf, X, y_true, display_labels=['aprovados', 'reprovados']):
    # Faça as previsões
    y_pred = clf.predict(X)
    
    # Calcule a matriz de confusão
    matriz_confusao = confusion_matrix(y_true, y_pred)
    
    # Plote a matriz de confusão
    plt.figure(figsize=(8, 6))
    sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=display_labels, yticklabels=display_labels)
    plt.title("Matriz de Confusão")
    plt.xlabel("Valores Preditos")
    plt.ylabel("Valores Reais")
    plt.show()

# Suponha que clf seja seu classificador de árvore de decisão
# Suponha que X_train e y_train sejam seus conjuntos de dados de treinamento
visualizar_matriz_confusao(clf, X_train, y_train)
````

````python
def calcular_acuracia(clf, X, y_true):
    # Faça as previsões
    y_pred = clf.predict(X)
    
    # Calcule a acurácia
    acuracia = accuracy_score(y_true, y_pred)
    
    return acuracia

acuracia_treinamento = calcular_acuracia(clf, X_train, y_train)
print('Acurácia na base de treinamento:', acuracia_treinamento)
````

#### 3) Vamos avaliar o modelo na base de testes

5. Como ficaria a acurácia se você classificasse todos os contratos como 'bons'?

````python
# 1) Classifique a base de teste de acordo com a árvore que você treinou no item 2.
def comparar_valores_previstos_e_reais(predicao_teste, y_teste):
    # Exiba os valores previstos
    print('Valores preditos:') 
    print(pd.Series(predicao_teste).value_counts(), '\n')

    # Exiba os valores reais
    print('Valores reais:')
    print(y_teste.value_counts())

# Suponha que predict_test e y_test sejam os valores preditos e verdadeiros, respectivamente
predict_test = clf.predict(X_test)
comparar_valores_previstos_e_reais(predict_test, y_test)
````

````python
# 2) Produza a visualização da matriz de confusão para a base de teste.

def visualizar_matriz_confusao_teste(clf, X_teste, y_teste, display_labels=['aprovados', 'reprovados']):
    # Faça as previsões
    predicoes_teste = clf.predict(X_teste)
    
    # Calcule a matriz de confusão
    matriz_confusao_teste = confusion_matrix(y_teste, predicoes_teste)
    
    # Plote a matriz de confusão
    plt.figure(figsize=(8, 6))
    sns.heatmap(matriz_confusao_teste, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=display_labels, yticklabels=display_labels)
    plt.title("Matriz de Confusão - Base de Teste")
    plt.xlabel("Valores Preditos")
    plt.ylabel("Valores Reais")
    plt.show()

# Suponha que clf seja seu classificador de árvore de decisão
# Suponha que X_test e y_test sejam seus conjuntos de dados de teste
visualizar_matriz_confusao_teste(clf, X_test, y_test)

````

````python
# 3) Calcule a acurácia da base de teste. Compare com a acurácia da base de treinamento.

def calcular_e_comparar_acuracias(clf, X_train, y_train, X_test, y_test):
    # Faça as previsões na base de treinamento e teste
    predicoes_train = clf.predict(X_train)
    predicoes_test = clf.predict(X_test)
    
    # Calcule as acurácias da base de treinamento e teste
    acuracia_train = accuracy_score(y_true=y_train, y_pred=predicoes_train)
    acuracia_test = accuracy_score(y_true=y_test, y_pred=predicoes_test)
    
    # Exiba as acurácias
    print('Acurácia da base de treinamento:', acuracia_train)
    print('Acurácia da base de teste:', acuracia_test)

# Suponha que clf seja seu classificador de árvore de decisão
# Suponha que X_train, y_train, X_test e y_test sejam seus conjuntos de dados de treinamento e teste
calcular_e_comparar_acuracias(clf, X_train, y_train, X_test, y_test)
````

````python
# 4) Treine uma nova árvore com número mínimo de observações por folha de 5 e máximo de profundidade de 10. 
#    Use o random_state = 123. Avalie a matriz de classificação. 
#    Observe a distribuição da predição - qual a proporção de proponentes foram classificados como 'maus'?

def plot_confusion_matrices(clf, X_test, y_test):
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
    fig.subplots_adjust(wspace=0.5)

    normalize_options = [None, 'true', 'pred', 'all']
    titles = ["No normalization", "Normalized by true values",
              "Normalized by predicted values", "Normalized by all values"]

    for i, norm in enumerate(normalize_options):
        ax = axes[i // 2, i % 2]
        disp = ConfusionMatrixDisplay.from_estimator(estimator=clf,
                                                      X=X_test,
                                                      y=y_test,
                                                      display_labels=['bons', 'maus'],
                                                      normalize=norm,
                                                      ax=ax)
        ax.set_title(titles[i])

    plt.show()

# Inicializando o classificador com os parâmetros especificados
nova_clf = DecisionTreeClassifier(max_depth=10, min_samples_leaf=5, random_state=123)

# Treinando o modelo
nova_clf.fit(X=X_train, y=y_train)

# Plotando as matrizes de confusão
plot_confusion_matrices(nova_clf, X_test, y_test)
````

````python
def calcular_proporcao_maus(n_maus, n_bons, n_total):
    """
    Calcula a proporção de proponentes classificados como 'maus'.
    
    Args:
    - n_maus (int): Número de proponentes classificados como 'maus'.
    - n_bons (int): Número de proponentes classificados como 'bons'.
    - n_total (int): Número total de proponentes.
    
    Returns:
    - float: Proporção de proponentes classificados como 'maus'.
    """
    return (n_maus / n_total) * 100

# Números de proponentes classificados como 'maus' e 'bons'
n_maus = 7
n_bons = 1937
n_total = n_maus + n_bons + 87  # Total de proponentes

# Calcula a proporção de proponentes classificados como 'maus'
proporcao_maus = calcular_proporcao_maus(n_maus, n_bons, n_total)

# Imprime a proporção com duas casas decimais
print("Proporção de proponentes classificados como 'maus':", round(proporcao_maus, 2), '%')
````

````python
# 5) Como ficaria a acurácia se você classificasse todos os contratos como 'bons'?

def calcular_acuracia(y_true, y_pred):
    """
    Calcula a acurácia dada as classificações verdadeiras e previstas.
    
    Args:
    - y_true (pandas.Series): Séries contendo os valores verdadeiros.
    - y_pred (pandas.Series): Séries contendo os valores previstos.
    
    Returns:
    - float: Acurácia do modelo.
    """
    return accuracy_score(y_true, y_pred)

# Valores reais
print("Valores reais:")
print(pd.Series(y_test).value_counts(), '\n')

# Todos os contratos da base teste classificados como 'bons'
y_pred_all_good = y_test.replace(to_replace=True, value=False)
print("Todos os contratos da base teste classificados como 'bons':")
print(pd.Series(y_pred_all_good).value_counts(), '\n')

# Calcula e imprime a acurácia
acuracia = calcular_acuracia(y_test, y_pred_all_good)
print('Acurácia:', round(acuracia * 100, 2), '%')
````
