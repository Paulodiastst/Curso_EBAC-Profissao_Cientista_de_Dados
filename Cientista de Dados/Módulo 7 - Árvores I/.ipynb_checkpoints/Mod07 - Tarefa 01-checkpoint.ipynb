#### 1) Carregue a base e avalie:

- As variáveis
- Tipos de dados de cada variável
- Quantidade de missings
- Distribuição da variável resposta (mau)

````python
# Carrega o arquivo CSV
df = pd.read_csv('demo01.csv')

# Exibe informações sobre o DataFrame antes da remoção de duplicatas
total_linhas, total_colunas = df.shape
print(f'Quantidade total de linhas: {total_linhas}')
print(f'Quantidade total de colunas: {total_colunas}')

# Identifica e exibe o número de linhas duplicadas
linhas_duplicadas = df.duplicated().sum()
print(f'Número de linhas duplicadas: {linhas_duplicadas}')

# Remove as linhas duplicadas
df.drop_duplicates(inplace=True)

# Exibe informações sobre o DataFrame após a remoção de duplicatas
linhas_removidas = total_linhas - len(df)
print(f'\nQuantidade de linhas após remoção: {len(df)} (removidas {linhas_removidas} linhas duplicadas)')

# Redefine o índice do DataFrame
df.reset_index(drop=True, inplace=True)

# Exibe o DataFrame resultante
print('\nDataFrame final:')
df
````

````python
# Exibe informações resumidas sobre o DataFrame
df_info = df.info()

# Separação visual entre as informações do DataFrame e os dados
print('\n' + '=' * 40 + '\n')

# Exibe as informações formatadas de uma maneira mais clara
print("Resumo das Informações do DataFrame:")
print(df_info)

# Exibir algumas estatísticas básicas sobre os dados
print('\nEstatísticas Básicas:')
print(df.describe())
````

````python
# Exibe um cabeçalho informativo
print("Colunas do DataFrame:")

# Itera sobre as colunas e imprime uma a uma
for col in df.columns:
    print(f'- {col}')

# Imprimi o número total de colunas
print(f'\nNúmero total de colunas: {len(df.columns)}')
````

````python
# Calcula a quantidade de valores ausentes em cada coluna
missing_values = df.isna().sum()

# Exibe um cabeçalho informativo
print("Quantidade de valores ausentes por coluna:")

# Itera sobre as colunas e imprime a quantidade de valores ausentes
for col, missing_count in missing_values.items():
    print(f'{col}: {missing_count}')

# Imprimir o número total de valores ausentes no DataFrame
total_missing = missing_values.sum()
print(f'\nNúmero total de valores ausentes no DataFrame: {total_missing}')
````

````python
# Exibe a distribuição da variável resposta 'mau'
print("Distribuição da variável resposta (mau):")
mau_distribution = df['mau'].value_counts()
print(mau_distribution)

# Adiciona uma linha visual de separação
print('\n' + '=' * 40 + '\n')

# Exibe a distribuição percentual da variável resposta 'mau'
print("Distribuição percentual da variável resposta (mau):")
mau_percentage_distribution = df['mau'].value_counts(normalize=True) * 100
print(mau_percentage_distribution)
````
