##### Análise básica para explorar a coluna Idade da Mãe ('IDADEMAE'), para entender a distribuição das idades das mães nos dados.

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

##### O APGAR é um sistema de avaliação rápida da saúde do recém-nascido, realizado no primeiro e quinto minutos após o nascimento. Ele fornece uma pontuação baseada em cinco critérios: Aparência, Pulso, Gesticulação, Atividade Muscular e Respiração. A pontuação total varia de 0 a 10, e valores mais altos indicam uma condição de saúde mais favorável.

```python
def categorize_apgar5(apgar5):
    if apgar5 <= 3:
        return 'asfixia grave'
    elif apgar5 <= 5:
        return 'asfixia moderada'
    elif apgar5 <= 7:
        return 'asfixia leve'
    else:
        return 'normal'

df_sinasc['APGAR5_categorizado'] = df_sinasc['APGAR5'].apply(categorize_apgar5)

print(df_sinasc['APGAR5_categorizado'].value_counts())
```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
# Defini a ordem desejada para o tempo de gestação
ordem_gestacao = [
    'Menos de 22 semanas', 
    '22 a 27 semanas', 
    '28 a 31 semanas', 
    '32 a 36 semanas', 
    '37 a 41 semanas', 
    '42 semanas e mais'
]

# Filtra o DataFrame para incluir apenas as categorias desejadas
df_filtrado = df_sinasc[df_sinasc['GESTACAO'].isin(ordem_gestacao)]

# Configura o estilo do Seaborn para melhorar a estética dos gráficos
sns.set(style="whitegrid")

# Cria subplots para cada categoria de tempo de gestação
plt.figure(figsize=(15, 10))
for i, gestacao in enumerate(ordem_gestacao, 1):
    plt.subplot(2, 3, i)
    sns.countplot(x='APGAR5', hue='SEXO', data=df_filtrado[df_filtrado['GESTACAO'] == gestacao])
    plt.legend(loc='upper left')
    plt.title('Tempo de gestação: {}'.format(gestacao))

# Ajusta o layout e exibir os subplots
plt.tight_layout()
plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Defini a ordem desejada para o tempo de gestação e categorias APGAR5
ordem_gestacao = [
    'Menos de 22 semanas', 
    '22 a 27 semanas', 
    '28 a 31 semanas', 
    '32 a 36 semanas', 
    '37 a 41 semanas', 
    '42 semanas e mais'
]

categorias_apgar5 = ['asfixia grave', 'asfixia moderada', 'asfixia leve', 'normal']

# Cria a tabela cruzada diretamente com as ordens definidas
tabela_cruzada = pd.crosstab(index=df_sinasc['GESTACAO'], 
                             columns=df_sinasc['APGAR5_categorizado'],
                             rownames=['Tempo de Gestação'],
                             colnames=['Categoria APGAR5'],
                             dropna=False)

# Reordena as linhas e colunas de acordo com as listas predefinidas
tabela_cruzada = tabela_cruzada.reindex(index=ordem_gestacao, columns=categorias_apgar5)

# Exibi a tabela cruzada
tabela_cruzada
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
# Calcula a tabela percentual dividindo cada valor na tabela cruzada pela soma das linhas
pct_tabela_cruzada = tabela_cruzada.div(tabela_cruzada.sum(axis=1), axis=0)

# Imprimi a soma percentual de cada linha para verificar se as percentagens somam 100%
print(pct_tabela_cruzada.sum(axis=1))

# Exibi a tabela percentual resultante
pct_tabela_cruzada
```

####  6) Os médicos costumam classificar parto prematuro quando a gestação tem menos que 37 semanas. Faça um gráfico de barras da tabela cruzada acima e conclua se você acha que este conjunto de dados corrobora ou não com a afirmação dos médicos.

```python
# Configura o tamanho padrão das figuras
plt.rc('figure', figsize=(12, 6))

# Cria um gráfico de barras para a tabela percentual
ax = pct_tabela_cruzada.plot.bar(color=['red', 'orange', 'yellow', 'green'], rot=45)

# Configura rótulos e legenda
plt.xlabel('Tempo de Gestação')
plt.ylabel('Percentual')
plt.title('Distribuição do Apgar no Quinto Minuto por Tempo de Gestação')
plt.legend(title='Apgar no quinto minuto', bbox_to_anchor=(1.05, 1), loc='upper left')

# Exibi os valores nas barras do gráfico
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2%}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

# Exibi o gráfico
plt.show()
```

- A análise revela a distribuição percentual das categorias APGAR no quinto minuto em relação ao tempo de gestação. O gráfico de barras destaca a predominância da categoria 'normal' para a maioria dos recém-nascidos, indicando uma condição de saúde favorável. Observa-se variação nas categorias de asfixia em diferentes períodos de gestação, com um aumento significativo nas gestações com menos de 22 semanas, sugerindo desafios significativos na saúde dos recém-nascidos nesses casos. Este insight é valioso para profissionais de saúde e pesquisadores, fornecendo informações cruciais para melhorias nos cuidados neonatais, especialmente em gestações precoces. A visualização clara e informativa do gráfico facilita a interpretação dos dados e a tomada de decisões informadas.

#### 7) Veja se há relação entre APGAR5 e o peso do bebê. Isso pode ser feito de diversas formas. Fique à vontade para escolher a melhor maneira na sua opinião, ou até testar mais de um dos gráficos que vimos. Fique à vontade também para consultar a documentação e utilizar algum outro recurso se achar conveniente.

```python
# Configura o tamanho padrão das figuras
plt.rc('figure', figsize=(12, 6))

# Cria subplots com duas colunas
fig, axes = plt.subplots(1, 2)

# Gráfico de barras usando seaborn
sns.barplot(x='APGAR5', y='PESO', data=df_sinasc, ax=axes[0])
axes[0].set_title('Distribuição de Peso por Categoria APGAR5')

# Gráfico de caixas (boxplot) usando seaborn
sns.boxplot(x='APGAR5', y='PESO', data=df_sinasc, ax=axes[1])
axes[1].set_title('Boxplot de Peso por Categoria APGAR5')

# Ajusta o layout para evitar sobreposição
plt.tight_layout()

# Exibi os gráficos
plt.show()
```

A análise dos gráficos de barras e boxplot revela padrões importantes na relação entre a pontuação APGAR no quinto minuto e o peso dos recém-nascidos.

Os gráficos indicam que a categoria 'normal' de APGAR está associada a pesos mais elevados, sugerindo uma correlação positiva entre uma condição de saúde favorável no quinto minuto após o nascimento e um peso mais alto. Em contraste, as categorias relacionadas à asfixia (grave, moderada, leve) exibem médias de peso decrescentes, indicando uma possível associação entre complicações no nascimento e pesos mais baixos.

O boxplot destaca que recém-nascidos com APGAR 'normal' tendem a ter uma distribuição de peso mais concentrada, enquanto as categorias associadas à asfixia mostram maior variabilidade. Essa variabilidade sugere diferentes níveis de gravidade nas condições de saúde dos recém-nascidos.

Esses insights podem orientar profissionais de saúde na identificação de padrões e na implementação de estratégias de cuidados neonatais, especialmente para casos com pontuações APGAR mais baixas, indicando a importância da vigilância e intervenções adequadas para garantir a saúde dos recém-nascidos.
