# **Como avaliar o Apgar do RN?**

O escore de Apgar é um sistema de avaliação da vitalidade do recém-nascido, realizado no primeiro, quinto e décimo minuto de vida. Ele é composto por cinco parâmetros:

* Aparência: cor da pele (coração)
* Pulso: frequência cardíaca (pulmão)
* Reflexo de choro: atividade muscular (músculos)
* Reatividade da musculatura dos membros (músculos)
* Respiração: movimentos respiratórios (pulmão)

Cada parâmetro é avaliado de 0 a 2 pontos, totalizando uma pontuação máxima de 10. Uma pontuação de 7 a 10 é considerada normal, enquanto uma pontuação inferior indica vitalidade deprimida.

O escore de Apgar no quinto minuto é importante para avaliar a resposta do recém-nascido às manobras de reanimação neonatal, se necessário. Ele também é um preditor de mortalidade neonatal e de sequelas neurológicas, como paralisia cerebral.

De acordo com estudos, um escore de Apgar inferior a 7 no quinto minuto está associado a um aumento do risco de mortalidade neonatal na primeira semana de vida, de 2 a 3 vezes. Além disso, um escore de Apgar inferior a 7 no quinto minuto também está associado a um aumento do risco de paralisia cerebral, de 2 a 3 vezes.

É importante ressaltar que o escore de Apgar é apenas uma ferramenta de avaliação. Ele não substitui o diagnóstico e o tratamento médico.

No Brasil, o escore de Apgar é usado como um dos critérios para a triagem neonatal de prematuridade e doenças congênitas.

**Interpretação do escore de Apgar no quinto minuto**

* 0 a 3: Asfixia grave
* 4 a 6: Asfixia moderada
* 7 a 10: Boa vitalidade, boa adaptação

**Fatores que podem afetar o escore de Apgar no quinto minuto**

* Prematuridade
* Asfixia perinatal
* Doenças congênitas
* Desvios anatômicos
* Idade materna avançada
* Tabagismo materno
* Uso de drogas ilícitas durante a gravidez
* Infecção materna

# **Semanas de gestação**

O gráfico mostra o número de mulheres grávidas por semanas de gestação. O eixo horizontal representa as semanas de gestação, de 0 a 42, e o eixo vertical representa o número de mulheres grávidas.

O número de mulheres grávidas aumenta gradualmente até a 37ª semana de gestação. Após a 37ª semana, o número de mulheres grávidas diminui.

A maioria das mulheres grávidas tem um parto a termo, entre 37 e 41 semanas. No entanto, uma pequena porcentagem de mulheres grávidas tem um parto prematuro, antes de 37 semanas.

```python
# Crie um gráfico de barras
counts = df_sinasc_RO['ESTCIVMAE'].value_counts().sort_index()
colors = ['#0000FF', '#FF0000', '#00FF00', '#FFFF00', '#999999']
bars = plt.bar(counts.index, counts, color=colors)

# Crie um objeto ScalarMappable para associar à barra de cores
sm = cm.ScalarMappable(cmap="brg", norm=mcolors.Normalize(vmin=counts.min(), vmax=counts.max()))
sm.set_array([])

# Adicione uma barra de cores
plt.colorbar(sm, label="Código", cmap="brg")

# Adicione uma legenda
plt.title("Distribuição do estado civil das mães de bebês nascidos em Rondônia em 2019", fontsize=18)
plt.xlabel("Estado civil", fontsize=14)
plt.ylabel("Número de mães", fontsize=14)
plt.legend(["Solteira", "Casada", "Viúva", "Separada judicialmente/Divorciada", "União consensual", "Ignorado"], loc='upper left')

# Adicione uma nota de rodapé abaixo do gráfico
plt.annotate(
    "1: Solteira\n2: Casada\n3: Viúva\n4: Separado judicialmente/Divorciado\n5: União consensual\n9: Ignorado",
    xy=(0, -0.4),
    fontsize=10,
    xycoords='axes fraction',
    horizontalalignment='left',
    verticalalignment='bottom'
)

# Exiba o gráfico
plt.show()
```

O gráfico mostra que a categoria "Solteira" é a mais comum, com porcentagem maior das mães. A categoria "Casada" é a segunda mais comum, com porcentagem um pouco menor das mães. As categorias "Viúva" e "União consensual" são menos comuns, respectivamente.

E bem provável que essa distribuição é semelhante às tendências em outros estados brasileiros. Em geral, a proporção de mães solteiras está aumentando em todo o Brasil, enquanto a proporção de mães casadas está diminuindo.
